```python
open_aging_df= df_open.copy()
```

```python
import pandas as pd
from datetime import datetime

# Ensure DATE column is datetime
open_aging_df['DATE'] = pd.to_datetime(open_aging_df['DATE'])

# Calculate age in days
today = pd.Timestamp.today()
open_aging_df['Age_Days'] = (today - open_aging_df['DATE']).dt.days

# Define age buckets
bins = [0, 15, 30, 60, 90, 180, float('inf')]
labels = ['<15Days', '16-30Days', '31-60Days', '61-90Days', '91-180Days', '>180Days']
open_aging_df['Age_Bucket'] = pd.cut(open_aging_df['Age_Days'], bins=bins, labels=labels, right=True)
```

```python
open_aging_df.head(1)
```

```python

# Pivot table with complaint type + age bucket
pivot_data = pd.pivot_table(
    open_aging_df,
    values='CLOSED/OPEN',          
    index=['COMPLAINT TYPE'],       # rows
    columns=['Age_Bucket'], # valid column names
    aggfunc='count',
    fill_value=0
)

# ➡️ Add Grand Total column (row-wise sum)
pivot_data['Grand_Total'] = pivot_data.sum(axis=1)

# ➡️ Add Grand Total row (column-wise sum)
pivot_data.loc['Grand_Total'] = pivot_data.sum(axis=0)


```
